```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('banking_data.csv')

# Strip any extra spaces from column names
df.columns = df.columns.str.strip()

# Task 1: Plot the total sum of Transaction_Amount per Account_Type using a bar plot
total_transaction_by_account = df.groupby('Account_Type')['Transaction_Amount'].sum()

# Create a bar plot
plt.figure(figsize=(8, 6))
total_transaction_by_account.plot(kind='bar', color='skyblue')
plt.title('Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Task 2: Create a pie chart to show the percentage of transactions per Branch
transactions_per_branch = df['Branch'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
transactions_per_branch.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title('Percentage of Transactions per Branch')
plt.ylabel('')  # Hide the y-label as it's not needed for a pie chart
plt.tight_layout()
plt.show()
```
