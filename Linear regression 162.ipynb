#### Fit a Linear regression model

```python
# Build regression model

import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data).fit()
```

```python
model1.summary()
```

```python
x = data["daily"].values
y = data["sunday"].values
plt.scatter(x,y,color = "m",marker = "o", s = 30)
b0 = 13.84
b1 = 1.33
y_hat = b0 + b1*x
plt.plot(x,y_hat, color = "g")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
newdata = pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
pred = model1.predict(data["daily"])
pred
```

```python
data["Y_hat"] = pred
data
```

```python
data["residuals"] = data["sunday"]-data["Y_hat"]
data
```

```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data).fit()
```

```python
model1.summary()
```

```python
mse = np.mean((data["daily"]-data["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE :",mse)
print("RMSE :",rmse)
```

```python
mae = np.mean(np.abs(data["daily"]-data["Y_hat"]))
mae
```

#### Checking the model rediuals of scatter plot and normality

```python
plt.scatter(data["Y_hat"],data["residuals"])
plt.xlable("Y_hat")
plt.ylable("Residuals")
```

## Observations :

#### . There appers to be no trend and the residuals are randomly placed around the zero error line 

##### . Hence the assumtion of homoscedasticty (constant varience in residuals) is satisfied.

```python
import statsmodels.api as sm
sm.qqplot(data["residuals"], line='45', fit=True)
plt.show()
```

```python
sns.histplot(data["residuals"],kde =True)
```

### Observations:
#### . The data points are seen to closely follow the reference line of normality.
#### . Hence the residuals are approximately normally distrubyted as also can be seen from the kde distuubution.
#### . All the assumptions os simple linear regression are verified and found to be valid in this case.
